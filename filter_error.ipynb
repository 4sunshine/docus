# Error filter

```python
N_resriction_files = 0
for f in os.listdir(result_root):
    if "NoRestriction" not  in f:
        N_resriction_files += 1

N_resriction_files
```

```python
n_errors_detected = 0 
p = '[\d]+[.,\d]+|[\d]*[.][\d]+|[\d]+'
data = {}
ID = 2 # 0 - прибор, 1 - единица измерения, 2 - погрешность

for f in os.listdir(result_root):
    with open(f'{result_root}/{f}', 'r') as fp:
        json_info = json.load(fp)
        
        # Error filter
        if len(json_info) > 2 and len(json_info[ID]['annotations']):
            splt = "".join(list(json_info[ID]['annotations'].values())).split("±")
            
            if len(splt) >= 2 and len([i[0] for i in re.finditer(p, splt[-1])]) == 1:
                print("-".join(f.split("-")[1:])[:-5], "±"+re.sub('[^0-9,]+', '', splt[-1]))
                data[str("-".join(f.split("-")[1:])[:-5])] = "±"+re.sub('[^0-9,]+', '', splt[-1])
                n_errors_detected += 1            
n_errors_detected
```

```python
with open('error.json', 'w') as f:
    json.dump(data, f)
```
